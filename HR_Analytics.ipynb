## Inference From The Data

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='numberOfProjects',hue='left',palette=['m','g'])
plt.xlabel('Number Of Projects')
plt.ylabel('Count')
plt.show()
```

**Inference**

The people with project count with 2, 6 and 7 are leaving most.

**Suggestion**

All the employees should be assigned 3-5 projects on average.

It should be redistributed among employees who are working on two projects from 6 and 7


```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='promotionInLast5years',hue='left')
plt.xlabel('Promotions In Last 5 Years')
plt.ylabel('Count')
plt.show()
```

**Inference**

People who are not getting the promotion for the last 5 years are leaving the most.

**Suggestion**

Timely promotion for the employees

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='salary',hue='left')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

**Inference**

People with low and medium salary are leaving the most.

**Suggestion**

Give salary Hike/incentives/Bonus according to experience and work

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='workAccident',hue='left',palette='plasma')
plt.xlabel('Work Accident')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df,x='dept',hue='left',palette='Set1')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

**Inference**

People of Sales, Technical and Support department are leaving the most.

**Suggestion**

For sales people based on their targets give some incentives.

Give option to move to other departments.

Team Management

Taking outings, lunch, fun activities to reduce the stress level.


```python
plt.figure(figsize=(12,5))
sns.histplot(data=df,x='avgMonthlyHours',hue='left',element='step',palette='Set2')
plt.xlabel('avgMonthlyHours')
plt.ylabel('Count')
plt.show()
```

**Inference**

People with minimum working hours or maximum monthly hours leave the most.
That is there is either no work or a lot of work.

**Suggestion**

Distribute work from high working hour employees to ow working hour employees.

```python
total_employees=df.groupby('dept').size()
```

```python
left_employee=df[df['left']==1].groupby('dept').size()
```

```python
attrition_rate=(left_employee/total_employees)*100
```

```python
attrition_df=pd.DataFrame({'Total Employees':total_employees,'Left Employees':left_employee,'Attrition rate':attrition_rate}).fillna(0)
```

```python
attrition_df
```

```python
plt.figure(figsize=(10,6))
attrition_df['Attrition rate'].plot(kind='bar',color='skyblue')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(
        attrition_df['Attrition rate'],
        labels=attrition_df.index,autopct='%1.1f%%',colors=plt.cm.plasma(attrition_df['Attrition rate']/max(attrition_df['Attrition rate'])),
        startangle=140,wedgeprops={'edgecolor':'black'}
        )
plt.title("Attrition Rate percentage of each department compared by whole atrrition rate of the company")
plt.show()
```
